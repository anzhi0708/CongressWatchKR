```python
from os import popen, listdir
from collections import defaultdict
from tqdm import tqdm
from pprint import pp
from multiprocessing import Queue, Process
from unicodedata import normalize
import matplotlib.pyplot as plt
from dearaj import *
import time  # For debugging.


Q = Queue()

ANALYZER = "/Users/aj/Public/py/dearAJ/src/ajFemaleMPAnalyzer.py"
WORKERS = []

# plt.rcParams["font.family"] = ["NanumGothic"]
# plt.figure(figsize=(48, 8))


def analyze(year: int, queue: Queue=Q):
    
    DIRECTORY = f"{year}_json"
    result = defaultdict(int)  # Initializing dictionary, return it at the end of function.
    
    for json_record in tqdm(listdir(DIRECTORY)):
        target = f"{DIRECTORY}/{json_record}"
        with popen(f"python3 {ANALYZER} --file={target} describe", "r") as output:
            lines = output.read().strip().splitlines()
            for line in lines:
                try:
                    line = line.split(':')
                    name = nomalize("NFKC", line[0].strip())
                    count = int(line[1].strip())
                    result[name] += count
                except IndexError as e:
                    print(f"{line = }")
                    print(e)
                    print(f"[Year {year}] Seems NO female MP spoke: {target}")
                    continue
    queue.put({year: result})
    
    return result
```

```python
worker2016 = Process(target=analyze, args=(2016,))
worker2017 = Process(target=analyze, args=(2017,))
worker2018 = Process(target=analyze, args=(2018,))
worker2019 = Process(target=analyze, args=(2019,))
worker2020 = Process(target=analyze, args=(2020,))
worker2021 = Process(target=analyze, args=(2021,))
```

```python
worker2016.start()
worker2017.start()
worker2018.start()
worker2019.start()
worker2020.start()
worker2021.start()

worker2016.join()
worker2017.join()
worker2018.join()
worker2019.join()
worker2020.join()
worker2021.join()
```
